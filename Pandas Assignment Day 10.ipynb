## Applying functions to columns and creating new columns

We need to create another column in d3, which contains  a boolean value for each city to indicate whether it's a union territory or not. 
- HINT: Chandigarh, Pondicherry and Delhi are only 3 union territories here. 


```python
# write function here

def is_ut(city):                     
#     # write code below     
    ut_cities = ['chandigarh', 'delhi', 'pondicherry']      
    return city.lower() in ut_cities      
     
d1['is_ut'] = d1['city'].apply(is_ut)                     
print(d1)  # Display d1 with the new 'is_ut' column indicating if the city is a union territory or not                                            
```

## Groupby

```python
# Create a dataframe using dictionary of your choice

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
      'Age': [24, 27, 22, 32], 'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']}  
df = pd.DataFrame(data)  


```

```python
# Use Groupby of single column with aggregate sum()
sum_age = df.groupby('City')['Age'].sum()
print(sum_age)  # Display the sum of ages for each city

```

```python
# Use Groupby of single column with aggregate count()
count_age = df.groupby('City')['Age'].count()
print(count_age)  # Display the count of ages for each city
```

```python
# Use Groupby of single column with aggregate min() and max()
min_age = df.groupby('City')['Age'].min()
max_age = df.groupby('City')['Age'].max()
print("Minimum ages by city:\n", min_age)  # Display the minimum age for each city
print("Maximum ages by city:\n", max_age)  # Display the maximum age for each city  

```

```python
# Use Groupby of any 2 columns with aggregate mean()
mean_age = df.groupby(['City', 'Name'])['Age'].mean()
print(mean_age)  # Display the mean age for each city and name combination
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()
min_age_2 = df.groupby(['City', 'Name'])['Age'].min()
max_age_2 = df.groupby(['City', 'Name'])['Age'].max()
print("Minimum ages by city and name:\n", min_age_2)  # Display the minimum age for each city and name combination
print("Maximum ages by city and name:\n", max_age_2)  # Display the maximum age for each city and name combination

```
